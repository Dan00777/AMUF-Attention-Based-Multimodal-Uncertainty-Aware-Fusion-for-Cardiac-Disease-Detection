```python
# ================================
# UA-MHDD: Fusion-Boosted Pipeline with Progressive Attention Fusion & Inline Visualization
# ================================

# === [STEP 1] INSTALL + SETUP ===
!pip install -q torchaudio librosa scikit-learn matplotlib seaborn umap-learn tqdm

import os
SAVE_DIR = "/content/drive/MyDrive/UA-MHDD"
os.makedirs(SAVE_DIR + "/models", exist_ok=True)
os.makedirs(SAVE_DIR + "/plots", exist_ok=True)
if not os.path.exists('./cache'):
    os.makedirs('./cache')

import pickle
import torch
import torchaudio
import librosa
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import random
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import *
from torch import nn
import torch.nn.functional as F
from torch.optim import AdamW
from torch.optim.lr_scheduler import ReduceLROnPlateau
from tqdm import tqdm

sns.set(style="whitegrid")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# === [STEP 2] IMPORT KAGGLE DATASETS ===
from google.colab import files
files.upload()
!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json
!kaggle datasets download -d bjoernjostein/physionet-challenge-2016 -p ./data --unzip
!kaggle datasets download -d mondejar/mitbih-database -p ./data --unzip

# === [STEP 3] DATA LOADING & CHECKS ===
ref_paths = glob.glob('./data/training-*/REFERENCE.csv') + glob.glob('./data/validation/REFERENCE.csv')
pairs, labels = [], []
for p in ref_paths:
    df = pd.read_csv(p, header=None, names=["recording", "label"])
    folder = os.path.dirname(p)
    for _, row in df.iterrows():
        rec_id = row["recording"].strip().lower()
        y = 0 if int(row["label"]) == 1 else 1
        wav_path = os.path.join(folder, rec_id + ".wav")
        if os.path.exists(wav_path):
            pairs.append((wav_path, wav_path))
            labels.append(y)
        else:
            print(f"[WARN] File not found: {wav_path}")
if len(pairs) == 0:
    raise RuntimeError("No .wav files paired with labels found! Check your dataset extraction.")

# === [STEP 4] TRAIN/VAL SPLIT, CLASS WEIGHTS, SAMPLER ===
train_p, val_p, train_l, val_l = train_test_split(pairs, labels, test_size=0.2, random_state=42, stratify=labels)
class_weights = compute_class_weight('balanced', classes=np.unique(train_l), y=train_l)
sampler = WeightedRandomSampler(weights=[class_weights[l] for l in train_l], num_samples=len(train_l), replacement=True)

# === [STEP 5] STRONG AUGMENTATION FUNCTIONS ===
def augment_ecg(ecg):
    if random.random() < 0.5:
        ecg = ecg + 0.01 * torch.randn_like(ecg)
    if random.random() < 0.5:
        ecg = ecg * (0.85 + 0.3 * torch.rand(1))
    if random.random() < 0.5:
        shift = random.randint(-60, 60)
        ecg = torch.roll(ecg, shifts=shift, dims=1)
    if random.random() < 0.3:
        L = ecg.shape[1]
        start = random.randint(0, L//2)
        ecg[:, start:start+random.randint(20,60)] = 0
    if random.random() < 0.2:
        ecg = ecg.flip(1)
    return ecg

def augment_pcg(pcg):
    if random.random() < 0.5:
        pcg = pcg + 0.01 * torch.randn_like(pcg)
    if random.random() < 0.5:
        pcg = pcg * (0.85 + 0.3 * torch.rand(1))
    if random.random() < 0.3:
        mels = pcg.shape[2]
        start = random.randint(0, mels//2)
        pcg[:, :, start:start+random.randint(8,16)] = 0
    return pcg

# === [STEP 6] DATASET CLASS (with augment option) ===
class UAMHDDDataset(Dataset):
    def __init__(self, pairs, labels, ecg_len=2000, pcg_len=3000, ecg_sr=250, pcg_sr=1000, augment=False):
        self.pairs = pairs
        self.labels = labels
        self.ecg_len = ecg_len
        self.pcg_len = pcg_len
        self.ecg_sr = ecg_sr
        self.pcg_sr = pcg_sr
        self.augment = augment

    def preprocess_ecg(self, path):
        cache_path = f"./cache/{os.path.basename(path)}_ecg.pkl"
        if os.path.exists(cache_path):
            with open(cache_path, 'rb') as f:
                return pickle.load(f)
        signal, sr = torchaudio.load(path)
        signal = librosa.resample(signal[0].numpy(), orig_sr=sr, target_sr=self.ecg_sr)
        signal = np.pad(signal[:self.ecg_len], (0, max(0, self.ecg_len - len(signal))))
        signal = (signal - np.mean(signal)) / (np.std(signal) + 1e-5)
        tensor = torch.tensor(signal, dtype=torch.float32).unsqueeze(0)
        with open(cache_path, 'wb') as f:
            pickle.dump(tensor, f)
        return tensor

    def preprocess_pcg(self, path):
        cache_path = f"./cache/{os.path.basename(path)}_pcg.pkl"
        if os.path.exists(cache_path):
            with open(cache_path, 'rb') as f:
                return pickle.load(f)
        signal, sr = torchaudio.load(path)
        signal = librosa.resample(signal[0].numpy(), orig_sr=sr, target_sr=self.pcg_sr)
        signal = np.pad(signal[:self.pcg_len], (0, max(0, self.pcg_len - len(signal))))
        mel = librosa.feature.melspectrogram(y=signal, sr=self.pcg_sr, n_fft=256, hop_length=64, n_mels=96)
        db = librosa.power_to_db(mel, ref=np.max)
        db = (db - db.min()) / (db.max() - db.min())
        tensor = torch.tensor(db, dtype=torch.float32).unsqueeze(0)
        with open(cache_path, 'wb') as f:
            pickle.dump(tensor, f)
        return tensor

    def __getitem__(self, idx):
        ecg_path, pcg_path = self.pairs[idx]
        ecg = self.preprocess_ecg(ecg_path)
        pcg = self.preprocess_pcg(pcg_path)
        label = torch.tensor(self.labels[idx])
        if self.augment:
            ecg = augment_ecg(ecg)
            pcg = augment_pcg(pcg)
        return ecg, pcg, label

    def __len__(self):
        return len(self.labels)

train_dl = DataLoader(UAMHDDDataset(train_p, train_l, augment=True), batch_size=8, sampler=sampler)
val_dl = DataLoader(UAMHDDDataset(val_p, val_l, augment=False), batch_size=8)

# === [STEP 7] MODEL DEFINITIONS ===
class ConvTransformerECG(nn.Module):
    def __init__(self, d_model=96):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv1d(1, 32, 5, padding=2), nn.GELU(),
            nn.Conv1d(32, 64, 5, padding=2), nn.GELU(),
            nn.Conv1d(64, d_model, 5, padding=2)
        )
        self.norm = nn.LayerNorm(d_model)
        self.encoder = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=d_model, nhead=4, batch_first=True),
            num_layers=2
        )

    def forward(self, x):
        x = self.conv(x).permute(0, 2, 1)
        x = self.encoder(x)
        return self.norm(x.mean(dim=1))

class BayesianTransformer(nn.Module):
    def __init__(self, d_model=96, dropout=0.2):
        super().__init__()
        self.embed = nn.Linear(96, d_model)
        self.encoder = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=d_model, nhead=4, dropout=dropout, batch_first=True),
            num_layers=2
        )
        self.dropout = nn.Dropout(dropout)
        self.norm = nn.LayerNorm(d_model)

    def forward(self, x):
        if x.dim() == 4:
            x = x.squeeze(1).permute(0, 2, 1)
        x = self.embed(x)
        x = self.encoder(x)
        return self.norm(self.dropout(x.mean(dim=1)))

class CrossModalAttentionFusion(nn.Module):
    def __init__(self, d_model=96, n_heads=4, dropout=0.2):
        super().__init__()
        self.ecg_proj = nn.Linear(d_model, d_model)
        self.pcg_proj = nn.Linear(d_model, d_model)
        self.cross_attn = nn.MultiheadAttention(d_model, n_heads, batch_first=True, dropout=dropout)
        self.norm = nn.LayerNorm(d_model*2)
        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(d_model*2, 2)

    def forward(self, ecg_feat, pcg_feat):
        ecg_proj = self.ecg_proj(ecg_feat).unsqueeze(1)  # (B, 1, D)
        pcg_proj = self.pcg_proj(pcg_feat).unsqueeze(1)  # (B, 1, D)
        attn_ecg, _ = self.cross_attn(ecg_proj, pcg_proj, pcg_proj)
        attn_pcg, _ = self.cross_attn(pcg_proj, ecg_proj, ecg_proj)
        fused = torch.cat([attn_ecg.squeeze(1), attn_pcg.squeeze(1)], dim=1)
        fused = self.norm(fused)
        fused = self.dropout(fused)
        out = self.fc(fused)
        return out

class UA_MHDD(nn.Module):
    def __init__(self, d_model=96, n_heads=4, dropout=0.2):
        super().__init__()
        self.ecg_trans = ConvTransformerECG(d_model=d_model)
        self.pcg_trans = BayesianTransformer(d_model=d_model, dropout=dropout)
        self.fusion_head = CrossModalAttentionFusion(d_model=d_model, n_heads=n_heads, dropout=dropout)

    def forward(self, ecg, pcg):
        ecg_feat = self.ecg_trans(ecg)
        B, C, H, W = pcg.shape
        pcg_seq = pcg.view(B, H, W)
        pcg_feat = self.pcg_trans(pcg_seq.permute(0, 2, 1))
        return self.fusion_head(ecg_feat, pcg_feat)

# === [STEP 8] THRESHOLD TUNING FUNCTION ===
def find_best_threshold(y_true, y_prob):
    best_thr, best_f1 = 0.5, 0
    for t in np.linspace(0.1, 0.9, 100):
        preds = (y_prob > t).astype(int)
        f1 = f1_score(y_true, preds)
        if f1 > best_f1:
            best_f1 = f1
            best_thr = t
    return best_thr

# === [STEP 9] TRAINING FUNCTION with TEMPORAL ENSEMBLE ===
def train_model(model, name, epochs=30, n_ensemble=5):
    model.to(device)
    opt = AdamW(model.parameters(), lr=1e-4)
    scheduler = ReduceLROnPlateau(opt, patience=3, factor=0.5, verbose=True)
    best_f1, patience = 0.0, 7
    best_paths = []
    for e in range(epochs):
        model.train()
        total_loss, correct = 0, 0
        pbar = tqdm(train_dl, desc=f"{name} Epoch {e+1}")
        for ecg, pcg, y in pbar:
            ecg, pcg, y = ecg.to(device), pcg.to(device), y.to(device)
            if name == "ECGOnly":
                out = model(ecg)
            elif name == "PCGOnly":
                out = model(pcg.squeeze(1).permute(0, 2, 1))
            else:
                out = model(ecg, pcg)
            loss = F.cross_entropy(out, y)
            opt.zero_grad()
            loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            opt.step()
            total_loss += loss.item()
            correct += (out.argmax(1) == y).sum().item()
            pbar.set_postfix({"loss": loss.item()})
            del ecg, pcg, y, out
        acc = correct / len(train_dl.dataset)
        print(f"âœ… Epoch {e+1}: Acc={acc:.4f}, Loss={total_loss:.2f}")
        scheduler.step(total_loss)
        torch.cuda.empty_cache()
        # === Evaluation + Early Stopping ===
        y_true, y_prob = [], []
        model.eval()
        with torch.no_grad():
            for ecg, pcg, y in val_dl:
                ecg, pcg, y = ecg.to(device), pcg.to(device), y.to(device)
                if name == "ECGOnly":
                    out = model(ecg)
                elif name == "PCGOnly":
                    out = model(pcg.squeeze(1).permute(0, 2, 1))
                else:
                    out = model(ecg, pcg)
                y_true.append(y)
                y_prob.append(out.softmax(1)[:, 1])
                del ecg, pcg, y, out
            torch.cuda.empty_cache()
        y_true = torch.cat(y_true).cpu().numpy()
        y_prob = torch.cat(y_prob).cpu().numpy()
        best_thr = find_best_threshold(y_true, y_prob)
        y_pred = (y_prob > best_thr).astype(int)
        f1 = f1_score(y_true, y_pred)
        auc = roc_auc_score(y_true, y_prob)
        print(f"ðŸ“ˆ Val F1={f1:.4f}, AUC={auc:.4f}, BestThr={best_thr:.3f}")
        save_path = f"{SAVE_DIR}/models/{name}_epoch{e+1}.pth"
        if f1 > best_f1:
            best_f1 = f1
            torch.save(model.state_dict(), save_path)
            print(f"ðŸ’¾ Best model saved for {name} at epoch {e+1}")
            best_paths.append(save_path)
            patience = 7
            if len(best_paths) > n_ensemble:
                os.remove(best_paths.pop(0))
        else:
            patience -= 1
            if patience == 0:
                print("â›” Early stopping")
                break
    evaluate_ensemble(model, name, best_paths)

# === [STEP 10] ENSEMBLE EVALUATION ===
from sklearn.calibration import calibration_curve

def expected_calibration_error(y_true, y_prob, n_bins=10):
    prob_true, prob_pred = calibration_curve(y_true, y_prob, n_bins=n_bins, strategy='uniform')
    bin_counts, bin_edges = np.histogram(y_prob, bins=n_bins, range=(0,1))
    bin_indices = np.digitize(y_prob, bin_edges[:-1], right=True)
    accs = np.array([y_true[bin_indices == i].mean() if np.any(bin_indices == i) else 0 for i in range(1, n_bins+1)])
    confs = np.array([y_prob[bin_indices == i].mean() if np.any(bin_indices == i) else 0 for i in range(1, n_bins+1)])
    n = len(y_true)
    ece = np.sum(bin_counts / n * np.abs(accs - confs))
    return ece, prob_true, prob_pred

def predictive_entropy(probs):
    return -np.sum(probs * np.log(probs + 1e-12), axis=1)

def save_and_show_fig(fig, name, plot_type):
    out_png = os.path.join(SAVE_DIR, "plots", f"{name}_{plot_type}.png")
    out_pdf = os.path.join(SAVE_DIR, "plots", f"{name}_{plot_type}.pdf")
    fig.savefig(out_png, dpi=300, bbox_inches='tight')
    fig.savefig(out_pdf, bbox_inches='tight')
    plt.show()
    plt.close(fig)

def plot_all_results(name, y_true, y_prob, y_pred, entropy, prob_true=None, prob_pred=None):
    cm = confusion_matrix(y_true, y_pred)
    fig, ax = plt.subplots()
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=ax)
    ax.set_title(f"{name} Confusion Matrix")
    ax.set_xlabel("Predicted")
    ax.set_ylabel("Actual")
    save_and_show_fig(fig, name, "confusion_matrix")

    fpr, tpr, _ = roc_curve(y_true, y_prob)
    auc = roc_auc_score(y_true, y_prob)
    fig, ax = plt.subplots()
    ax.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
    ax.plot([0,1], [0,1], "--", color="gray")
    ax.set_xlabel("False Positive Rate")
    ax.set_ylabel("True Positive Rate")
    ax.set_title(f"{name} ROC Curve")
    ax.legend()
    save_and_show_fig(fig, name, "roc_curve")

    prec, rec, _ = precision_recall_curve(y_true, y_prob)
    fig, ax = plt.subplots()
    ax.plot(rec, prec, color='blue')
    ax.set_xlabel("Recall")
    ax.set_ylabel("Precision")
    ax.set_title(f"{name} Precision-Recall Curve")
    save_and_show_fig(fig, name, "pr_curve")

    fig, ax = plt.subplots()
    ax.plot(prob_pred, prob_true, marker='o', label="Model")
    ax.plot([0,1], [0,1], "--", color="gray", label="Perfectly calibrated")
    ax.set_xlabel("Confidence")
    ax.set_ylabel("Accuracy")
    ax.set_title(f"{name} Reliability Diagram")
    ax.legend()
    save_and_show_fig(fig, name, "reliability_diagram")

    fig, ax = plt.subplots()
    ax.hist(entropy, bins=30, color="purple", alpha=0.7)
    ax.set_xlabel("Predictive Entropy")
    ax.set_ylabel("Count")
    ax.set_title(f"{name} Predictive Entropy Histogram")
    save_and_show_fig(fig, name, "entropy_histogram")

    fig, ax = plt.subplots()
    ax.hist(y_prob, bins=30, color="green", alpha=0.7)
    ax.set_xlabel("Predicted Probability (class 1)")
    ax.set_ylabel("Count")
    ax.set_title(f"{name} Predicted Probability Histogram")
    save_and_show_fig(fig, name, "probability_histogram")

def evaluate_ensemble(model, name, checkpoint_paths):
    y_true, y_probs = [], []
    for ecg, pcg, y in val_dl:
        ecg, pcg, y = ecg.to(device), pcg.to(device), y.to(device)
        logits = []
        for ckpt in checkpoint_paths:
            model.load_state_dict(torch.load(ckpt))
            model.eval()
            with torch.no_grad():
                if name == "ECGOnly":
                    out = model(ecg)
                elif name == "PCGOnly":
                    out = model(pcg.squeeze(1).permute(0, 2, 1))
                else:
                    out = model(ecg, pcg)
                logits.append(out.softmax(1).cpu().numpy())
        logits = np.mean(logits, axis=0)
        y_true.append(y.cpu().numpy())
        y_probs.append(logits)
    y_true = np.concatenate(y_true)
    y_probs = np.concatenate(y_probs)
    y_prob = y_probs[:,1]
    best_thr = find_best_threshold(y_true, y_prob)
    y_pred = (y_prob > best_thr).astype(int)
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    auc = roc_auc_score(y_true, y_prob)
    f2 = fbeta_score(y_true, y_pred, beta=2)
    ece, prob_true, prob_pred = expected_calibration_error(y_true, y_prob)
    entropy = predictive_entropy(y_probs)
    mean_entropy = entropy.mean()
    pd.DataFrame({
        "accuracy": [acc], "precision": [prec], "recall": [rec], "f1": [f1], "f2":[f2], "auc": [auc],
        "ece": [ece], "mean_entropy": [mean_entropy]
    }).to_csv(f"{SAVE_DIR}/plots/{name}_ensemble_metrics.csv", index=False)
    plot_all_results(name, y_true, y_prob, y_pred, entropy, prob_true, prob_pred)
    print("\n\\begin{tabular}{lccccccc}")
    print("\\toprule")
    print("Model & Accuracy & Precision & Recall & F1 & F2 & AUC & ECE \\\\")
    print("\\midrule")
    print(f"{name} & {acc:.4f} & {prec:.4f} & {rec:.4f} & {f1:.4f} & {f2:.4f} & {auc:.4f} & {ece:.4f} \\\\")
    print("\\bottomrule")
    print("\\end{tabular}")
    print(f"Mean predictive entropy: {mean_entropy:.4f}")
    print(f"All plots saved in: {SAVE_DIR}/plots/ (PNG & PDF)")
    print(f"Best threshold for F1: {best_thr:.4f}")

# === [STEP 11] FINAL TRAINING AND EVALUATION ===
print("ðŸš€ Training ECGOnly Model")
ecg_model = ConvTransformerECG()
train_model(ecg_model, "ECGOnly")

print("\nðŸš€ Training PCGOnly Model")
pcg_model = BayesianTransformer()
train_model(pcg_model, "PCGOnly")

print("\nðŸš€ Training Fusion Model (Attention-Based Cross-Modal Fusion)")
fusion_model = UA_MHDD()
train_model(fusion_model, "Fusion")
```
